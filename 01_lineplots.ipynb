## Study the data of Iris with the help of Lineplots

|Step-1|
|------|


### Import these libararies

- seaborn
- matplotlib
- pandas
- numpy
- scipy

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

|Step-2|
|------|

### Load_dataset


```python
flower=sns.load_dataset("iris")
flower
```

|Step-3|
|------|
### Making a plot

```python
sns.lineplot(x="sepal_length",y="sepal_width",data =flower)
plt.show
```

### How to add a title?

```python
sns.lineplot(x="sepal_length",y="sepal_width",data =flower)
plt.title("flower Data")
plt.title
```

### How to limit a Plot?


```python
sns.lineplot(x="sepal_length",y="sepal_width",data =flower)
plt.xlim(2)
plt.ylim(1)
plt.show
```

### How to increase or decrease figure size ?

```python
plt.figure(figsize=(2,3))
sns.lineplot(x="sepal_length",y="sepal_width",data =flower)
plt.title("flower plot")
sns.set_style(style=None,rc=None)
sns.set_style("dark")
```
